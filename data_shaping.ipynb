```python
import pandas as pd
import yfinance as yf

tickers = ['BTC','ETH']
frames = []

for ticker in tickers:
    df = yf.download(tickers=ticker+'-GBP', period='ytd', interval='1d')
    df.sort_index(ascending=False, inplace=True)
    df.reset_index(inplace=True)
    df['Ticker'] = ticker
    frames.append(df)
    
df = pd.concat(frames)
df = pd.melt(df, id_vars=['Date','Ticker'])
```
